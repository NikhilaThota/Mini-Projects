## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas-docs.github.io/pandas-docs-travis/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# Inspect the coulmn 'mjtheme_namecode'
pd.set_option('max_colwidth',150)
jdft = jdf[['mjtheme_namecode']]
jdft
```

```python
# Only get the 'values' for further evaluation
jdft1 = jdft.to_json(orient = "values")
jdft1
```

```python
# Clean the data in json file
jdft1=jdft1.replace('[[','')
```

```python
# Clean the data in json file
jdft1=jdft1.replace(']]','')
```

```python
# Load the cleaned data in 'd'
d = json.loads(jdft1)
```

```python
# Create a dictionary to fill the missing 'name' values.
pt_dict = {}
for e in d:
        key = e['code']
        val = e['name']
        if key in pt_dict:
            if not pt_dict[key]:
                pt_dict[key] = val
        else:
            pt_dict[key] = val        
```

```python
# Dictionary with all the 'code' and 'name' pairs
pt_dict
```

```python
# Using pt_dict{} as reference fill the missing 'name' values in d
for e in d:
    if not e['name']:
        e['name'] = pt_dict[e['code']]
```

```python
# Create new DataFrame df to be filled with the 'code', 'name' values
df = pd.DataFrame.from_dict(d, orient='columns')
```

```python
df
```

```python
# Top 10 major project theme names
df['name'].value_counts().head(10)
```
